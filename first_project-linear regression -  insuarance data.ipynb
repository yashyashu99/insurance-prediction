```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("insurance.csv")
```

```python
data.head()
```

# Data Preprocessing


```python
# One Hot Encoding
data["sex"]=data["sex"].apply({'female':1,'male':0}.get)
```

```python
data.head()
```

```python
data["smoker"]=data["smoker"].apply({"yes":1,"no":0}.get)
data["region"]=data["region"].apply({"southwest":0,"southeast":1,"northwest":2,"northeast":4}.get)
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

# Dividing The Data Into Dependent and Independent

```python
x=data[["age","sex","bmi","children","smoker","region"]] # independent varibales

y= data[["expenses"]]  # dependent varibales
```

```python
x.head()
```

```python
y.head()
```

# Spliting the data into training and testing

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)  # 0.2=20% data goes for the test and 80% goes for training
```

# MLM using Linear Regression Algorithm

```python
from sklearn.linear_model import LinearRegression   # linear_model == module

regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
y_predict=regression.predict(x_test)
```

```python
print(y_predict)  # it id predicted values or data
```

```python
regression.score(x,y)  # accuracy of our model 
```
